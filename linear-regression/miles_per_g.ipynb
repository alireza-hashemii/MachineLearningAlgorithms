```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
```

```python
df = sns.load_dataset('mpg')
```

```python
df = df.dropna(axis=0, how='any')
```

```python
x = df['horsepower']
y = df['mpg']
```

```python
split_ratio = 0.7
split_index = int(len(x) * split_ratio)

x_train = x[:split_index].to_numpy()
y_train = y[:split_index]

x_test = x[split_index:].to_numpy()
y_test = y[split_index:]
```

```python
x_train_reshaped = x_train.reshape(-1, 1)
x_test_reshaped = x_test.reshape(-1, 1)
```

```python
model = LinearRegression()
model.fit(x_train_reshaped, y_train)
```

```python
print(f'Slope is -> {model.coef_[0]}')
print(f"Y-intercept is -> {model.intercept_}")
```

```python
plt.scatter(x_train, y_train, label="Training Data")
plt.scatter(x_test, y_test, label="Testing Data", color='r')
plt.plot(x_train, model.predict(x_train_reshaped), color='blue', label=f'{model.coef_[0]:.3f} X + {model.intercept_:.3f}' )
plt.legend()
plt.show()
```

```python
y_pred = model.predict(x_test_reshaped)
```

```python
# mae
mae_rate = mean_absolute_error(y_test, y_pred)
mae_rate
```

```python
# mse
mse_rate = mean_squared_error(y_test, y_pred)
mse_rate
```

```python
rmse_rate = np.sqrt(mse_rate)
rmse_rate
```
