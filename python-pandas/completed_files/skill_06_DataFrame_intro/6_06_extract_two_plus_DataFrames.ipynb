# Extract two or more columns from a DataFrame
Similar to filtering tool in spreadsheets

### Import Pandas

```python
import pandas as pd
```

### Read CSV

```python
salaries = pd.read_csv("mls_salaries.csv")
```

```python
# called the head method to see what were working with
salaries.head()
```

### Extracting Two Columns

```python
# pass in a list to return two columns as DataFrame
salaries[["club", "last_name"]]
```

```python
# pass in a list to return two columnns as DataFrame REVERSED!
salaries[["last_name","club"]]
# salaries[["last_name","club"]].tail(3)
```

### Extracting Three Columns

```python
# pass in a list to return more than two columnns as DataFrame with CUSTOM ORDERING!
salaries[["guaranteed_compensation","last_name","club"]]
# salaries[["guaranteed_compensation","last_name","club"]].tail(3)
```

### Refactor Using a Variable

```python
# best practice: refactor your code for readabililty
# save the list as  a variable 
salaries_custom = ["guaranteed_compensation","last_name","club"]
salaries[salaries_custom]
```
