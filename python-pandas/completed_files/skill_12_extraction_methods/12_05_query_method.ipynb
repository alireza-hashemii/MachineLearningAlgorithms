# `.query()` method

### Import pandas

```python
import pandas as pd
```

### Read CSV and set rating index sorted by labels

```python
cereal = pd.read_csv("cereal.csv", index_col="name")
cereal.sort_index(inplace=True)
cereal
```

```python
# check out the query method documentation first
# cereal.query("mfr == 'K'") quote type agnostic
cereal.query('mfr == "K"')
```

```python
cereal.query('mfr != "K"')
```

```python
cereal.query("calories > 100")
```

```python
cereal.query('mfr == "K" and calories > 100')
```

```python
cereal.query('mfr == "K" or calories > 100')
```

```python
cereal.query('mfr in ["K", "G"]')
```

```python
cereal.query('mfr not in ["K", "G"]')
```
