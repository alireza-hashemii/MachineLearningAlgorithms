# `.nsmallest( )` & `nlargest( )` methods

### Import pandas

```python
import pandas as pd
```

### Read CSV and set rating index sorted by labels

```python
cereal = pd.read_csv("cereal.csv", index_col="name")
cereal.sort_index(inplace=True)
cereal
```

### Return rows with largest values in a columns

```python
# this is one way
cereal.sort_values("rating", ascending = False).head(3)
```

```python
# or you could use nlargest() method: see docs
cereal.nlargest(3, "rating")
```

### Return rows with smallest values in a columns

```python
# the long way
cereal.sort_values("rating", ascending = True).head(3)
```

```python
# you can also use nsmallest() method
cereal.nsmallest(3, "rating")
```

```python
cereal.nsmallest(n = 3, columns = "rating")
```
