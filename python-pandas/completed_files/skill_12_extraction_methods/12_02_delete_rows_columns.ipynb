# `.drop( )` method

### Import pandas

```python
import pandas as pd
```

### Read CSV and set rating index sorted by labels

```python
cereal = pd.read_csv("cereal.csv", index_col="name")
cereal.sort_index(inplace=True)
cereal
```

### Drop a row from a DataFrame

```python
# cereal.drop("All-Bran", inplace=True)
cereal.drop("All-Bran")
```

### Drop multiple rows

```python
# cereal.drop(["Trix", "Triples"], inplace=True)
cereal.drop(["Trix", "Triples"])
```

### Remove columns

```python
# cereal.drop(["mfr", "type"], axis = 1, inplace=True)
cereal.drop(["mfr", "type"], axis = "columns")
```

### Keeping it Pythonic

```python
# you can also use the python del keyword: no inplace needed
del cereal["calories"]
```
