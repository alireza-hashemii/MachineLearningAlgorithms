### Read CSV `.index_col()` parameter

```python
cereal = pd.read_csv("cereal.csv", index_col = "name")
cereal.head()
```

### `.set_index()` method

```python
cereal = pd.read_csv("cereal.csv")
cereal.head()
```

```python
cereal.set_index(keys = "name")
```

```python
# makes it permanent
cereal.set_index(keys = "name", inplace=True)
```

```python
cereal.head()
```

```python
# reset original index
cereal.reset_index()
```

### Reset original index

```python
# reset original index
cereal.reset_index(inplace=True)
```

```python
# name as index again, make it permanent
cereal.set_index(keys = "name", inplace=True)
cereal.head()
```

```python
# default is not to drop the 
cereal.reset_index(drop=False)
```

```python
cereal.reset_index(drop=True)
```
