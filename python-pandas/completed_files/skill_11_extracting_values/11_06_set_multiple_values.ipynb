# Set multiple values in a particular row

### Import pandas

```python
import pandas as pd
```

### Read CSV and set rating index sorted by labels

```python
cereal = pd.read_csv("cereal.csv", index_col="name")
cereal.sort_index(inplace=True)
cereal
```

### Set multiple values from "C" to "cereal"

```python
cereal_type = cereal["type"] == "C"
```

```python
cereal_type
```

```python
# don't do it this way ⚠️ Try using .loc[row_indexer,col_indexer] = value instead
cereal[cereal_type]["type"] = "cereal"
```

```python
cereal
```

```python
cereal.loc[cereal_type,"type"]
```
