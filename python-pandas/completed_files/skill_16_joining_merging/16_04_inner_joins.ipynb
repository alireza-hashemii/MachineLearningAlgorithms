# Inner Joins

### Import pandas

```python
import pandas as pd
```

### Read CSV

```python
customers = pd.read_csv("cafe_customers.csv")
items = pd.read_csv("cafe_items.csv")
week_1 = pd.read_csv("cafe_week_1_sales.csv")
week_2 = pd.read_csv("cafe_week_2_sales.csv")
```

### Perform an inner join on week_1 and week_2 sales

```python
# only returns customers who purchased in week_1 AND week_2
week_1.merge(week_2, how = "inner", on = "cust_id")
```

```python
# note the duplicate cust_ids
week_1.merge(week_2, how = "inner", on = "cust_id", suffixes = ["_wk_1", "_wk_2"])
```

```python
# note the unique cust_ids
week_1.merge(week_2, how = "inner", on = ["cust_id", "item_id"])
```
