# Inclusion `isin()` Method

### Import Pandas

```python
import pandas as pd
```

### Employee data


```python
# df = pd.read_csv("employee.csv", parse_dates = ["start_date"])
df = pd.read_csv("employee.csv")
```

### Head

```python
# called the head method to see what were working with
df.head()
```

### The `isin()` Method

```python
mask = df["team"].isin(["data analytics"])
mask
```

### Extract by Passing `mask` to `df[]`

```python
df[mask]
```

```python
mask = df["team"].isin(["data analytics","management","engineering"])
mask
```

```python
df[mask]
```

```python
mask_1 = df["team"].isin(["data analytics","management","engineering"])
mask_2 = df["remote"].isin([True])
```

```python
df[mask_1 & mask_2]
```

```python
df[mask_1 | mask_2]
```
