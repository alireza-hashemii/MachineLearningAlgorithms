# Drop Duplicates

### Import Pandas

```python
import pandas as pd
```

### Employee data


```python
df = pd.read_csv("employee.csv", parse_dates = ["start_date"])
```

### Head

```python
# called the head method to see what were working with
df.head()
```

```python
len(df)
```

```python
# all values must be duplicates to be dropped
len(df.drop_duplicates())
```

```python
# keeps the first duplicate instance and drops the rest
len(df.drop_duplicates(subset = ["first_name"], keep="first"))
```

```python
df.drop_duplicates(subset = ["first_name"], keep="first")
```

```python
len(df.drop_duplicates(subset = ["first_name"], keep="last"))
```

```python
df.drop_duplicates(subset = ["first_name"], keep="last")
```

```python
# counts all dropped dupliactes
len(df.drop_duplicates(subset = ["first_name"], keep=False))
```

```python
# drops all dupliactes 
df.drop_duplicates(subset = ["first_name"], keep=False)
```

```python
# see all deleted occurrences in two columns and make permanent
df.drop_duplicates(subset = ["first_name", "team"], inplace=True)
```
